```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
```

```python
df=pd.read_csv('epa-sea-level.csv')
```

```python
df = df.dropna(subset=['Year', 'CSIRO Adjusted Sea Level'])
r=linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
e=pd.Series([i for i in range(1875,2050)])
```

```python
a=df[df['Year']>=2000]
rn=linregress(a['Year'],a['CSIRO Adjusted Sea Level'])
x=pd.Series(i for i in range(2000,2050))
y=rn.slope*x+rn.intercept
```

```python
sns.scatterplot(data=df,x="Year",y="CSIRO Adjusted Sea Level")
plt.plot(e,r.slope*e+r.intercept,color='red')

plt.plot(x,y,color='blue')
```
